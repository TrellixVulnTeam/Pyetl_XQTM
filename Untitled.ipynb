```python
import pandas as pd
import os
```

```python
source_path = r'./res_gossiping'
file_list = os.listdir(source_path)
file_list
```

```python
df_list =[]
for article in file_list:
    with open(source_path + '/' + article, 'r' ,encoding = 'utf-8') as f:
        tem_str = f.read()
    try:
        info_list = tem_str.split('---split---')[1].split('\n')[1:7]
    except IndexError as e:
        print(e.args)
    df_list.append(info_list)
df_list
```

```python
df = pd.DataFrame(columns=['推','噓','分數','作者','標題','時間'])
df
```

```python
new_df = pd.DataFrame(df_list,columns=['推','噓','分數','作者','標題','時間'])
new_df
```

```python
df =df.append(new_df)
df
```

```python
def column_filter(tmpstr):
    return tmpstr.split(': ')[1]
```
